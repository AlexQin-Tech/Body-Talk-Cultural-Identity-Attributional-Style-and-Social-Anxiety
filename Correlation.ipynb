```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import spearmanr
import warnings
warnings.filterwarnings("ignore")

df = pd.read_excel(r"Form Data",
                   sheet_name="Data")

cols = ["ACis","ACac","BC","NFT","NMTa","NMTb","PBTi","PBTe","EB","PB","P","SI","IBI","PA"]
data = df[cols].dropna()
n = len(cols)

corr = np.zeros((n, n))
pval = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        if i == j:
            corr[i,j] = 1.0
            pval[i,j] = 0
        else:
            corr[i,j], pval[i,j] = spearmanr(data.iloc[:,i], data.iloc[:,j])

def sig(p):
    if p < 0.001: return "***"
    elif p < 0.01: return "**"
    elif p < 0.05: return "*"
    return ""

plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(18, 15))
cmap = plt.cm.RdBu_r
norm = plt.Normalize(-1, 1)
base_radius = 0.16
scale_factor = 0.36

for i in range(n):
    for j in range(n):
        r = corr[i, j]
        p = pval[i, j]
        
        if i == j:
            ax.text(j, i, cols[i], ha='center', va='center',
                    fontsize=25, fontweight='bold', color='black')
            continue
            
        if i > j:
            radius = base_radius + abs(r) * scale_factor
            circle = plt.Circle((j, i), radius, color=cmap(norm(r)), alpha=0.96, ec='none')
            ax.add_patch(circle)
            if p < 0.05:
                star_color = 'black' if abs(r) < 0.62 else 'white'
                ax.text(j, i, sig(p), ha='center', va='center',
                        fontsize=19, fontweight='bold', color=star_color)

        if i < j:

            ax.text(j, i, f'{r:.3f}', ha='center', va='center',
                    fontsize=17, fontweight='bold', color='black')

ax.set_xlim(-0.8, n-0.2)
ax.set_ylim(n-0.2, -0.8)
ax.set_xticks(np.arange(n))
ax.set_yticks(np.arange(n))
ax.set_xticklabels(cols, rotation=45, ha='right', fontsize=20, fontweight='bold')
ax.set_yticklabels(cols, fontsize=20, fontweight='bold')
ax.tick_params(length=0)
ax.set_aspect('equal')

sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = plt.colorbar(sm, ax=ax, shrink=0.82, pad=0.02)
cbar.set_label("Spearman's ρ", fontsize=22, fontweight='bold')

plt.tight_layout()
plt.savefig("Results",
            format='pdf', bbox_inches='tight', dpi=1200)
plt.show()

print("V1")
print("V2")
print("V3")
```
